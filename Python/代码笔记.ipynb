```python
import os
from PIL import Image
from numpy import *
from pylab import *#pylab中有imshow
img_prefix=['png']


def read_images(base_path):
    files = os.listdir(base_path)   #缩进要统一
    return files
    

def compute_average(imlist, base_path):
    averageim = array(Image.open(os.path.join(base_path, imlist[0])), 'f')#读取指定路径下文件
    for imname in imlist[1:]:
        try:
            averageim+=array(Image.open(os.path.join(base_path, imname)))#打开文件夹中所有图像方法
        except:
            print(imname + '...skipped')
    averageim/=len(imlist)
    return array(averageim, 'uint8')


base_path = "/home/xc/notebook/meitong"
files = read_images(base_path)
averageim = compute_average(files, base_path)
imshow(averageim)
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import imageio
u=np.linspace(0,2*3.141592653589793,100)
v=np.linspace(0,3.141592653589793,100)
a,b= 15000000,10000000
x=a*np.outer(np.cos(u),np.sin(v))
y=a*np.outer(np.sin(u),np.sin(v))
z=b*np.outer(np.ones(np.size(u)),np.cos(v))
fig = plt.figure()
ax=fig.add_subplot(111, projection='3d')
ax.plot_surface(x,y,z,cmap=plt.get_cmap('gnuplot'),alpha=0.7)
ax.view_init(270,0)
for i in np.linspace(0,2*3.141592653589793,100):
  x_yuanzhui=a*math.cos(math.radians(60))*math.sin(i)
  y_yuanzhui=a*math.cos(math.radians(60))*math.cos(i)
  z_yuanzhui=b-a*math.sin(math.radians(60))
  plt.plot([0,x_yuanzhui],[0,y_yuanzhui],[10000000,z_yuanzhui],color='black',linestyle='--')
filename="22"
plt.axis('off')
plt.savefig(filename)
```

```python
#分通道处理图像
from PIL import Image
from numpy import *
from scipy.ndimage import filters
from pylab import *
import matplotlib.pyplot as plt
im = array(Image.open('blank_billboard.jpg'))
im2=zeros(im.shape)
index=141
plt.subplot(index)
imshow(im)
for sigma in (2,5,10):  #双重for循环生成不同sigma的高斯滤波图像
    for i in range(3):
        im2[:,:,i] = filters.gaussian_filter(im[:,:,i],sigma)
    im2=uint8(im2)   #处理后需要转化成8进制
    index+=1      #直接加1创建子图
    plt.subplot(index)
    imshow(im2)
```
